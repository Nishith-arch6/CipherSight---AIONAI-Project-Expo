```python
import jwt
import datetime
import hashlib

# 1. SECRET KEY (In a real app, this is an environment variable)
SECRET_KEY = "ciphersight_super_secret_key"

def generate_secure_token(user_id, role="admin"):
    """Generates a timed token for the Full-Stack Lead's Login system."""
    payload = {
        'user_id': user_id,
        'role': role,
        'exp': datetime.datetime.utcnow() + datetime.timedelta(hours=2)
    }
    return jwt.encode(payload, SECRET_KEY, algorithm='HS256')

def verify_token(token):
    """Verifies the token before the Routing Lead can change a signal."""
    try:
        decoded = jwt.decode(token, SECRET_KEY, algorithms=['HS256'])
        return True, decoded
    except jwt.ExpiredSignatureError:
        return False, "Token expired. Please log in again."
    except jwt.InvalidTokenError:
        return False, "Unauthorized Access Detected!"

# TEST IT
token = generate_secure_token("Officer_Smith")
print(f"Generated Token: {token}")
```

```python
from cryptography.fernet import Fernet

# Generate and save a key (In Colab, we'll keep it in a variable)
key = Fernet.generate_key()
cipher_suite = Fernet(key)

def encrypt_location_data(lat, lon):
    message = f"{lat},{lon}".encode()
    return cipher_suite.encrypt(message)

def decrypt_location_data(encrypted_text):
    return cipher_suite.decrypt(encrypted_text).decode()

# TEST IT
secret_coords = encrypt_location_data(12.9716, 77.5946)
print(f"Encrypted Coords: {secret_coords}")
print(f"Decrypted Coords: {decrypt_location_data(secret_coords)}")
```
